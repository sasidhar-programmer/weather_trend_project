```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sb 
sb.set_style('darkgrid')
```

```python
temp_data = pd.read_csv('Temperature_data.csv') 
temp_data.head()
 
```

```python
temp_data.tail()
```

```python
temp_data.dropna(inplace=True) # dropping null vaules 
```

```python
temp_data.describe()  # statistical view of the data  
```

```python
glb_mv_avg = temp_data['city_temp'].rolling(10).mean()
local_mv_avg = temp_data['global_temp'].rolling(10).mean() 
sb.lineplot(y = glb_mv_avg, x = temp_data['year']) 
sb.lineplot(y = local_mv_avg, x = temp_data['year'])
plt.title('Moving average graph\n global vs city temperature') 
plt.ylabel('Temperature Â°c') 
```

```python

#sb.lineplot(x = temp_data['year'], y = temp_data['city_temp'], hue = temp_data['city'] ) 
sb.lineplot(data = temp_data, x = 'year', y = 'city_temp', hue = 'city',err_style='bars')
sb.lineplot(data = temp_data , x ='year', y = 'global_temp', hue ='country',err_style='bars')  
plt.title('Temperature Graph Comparison\n Hyderabad vs Global_temp')
plt.ylabel('Temperature Â°c')

```
